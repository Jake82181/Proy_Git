# **Agente Reactivo: Consejos de Bienestar Digital**

## **EJECUTAR AGENTE**

```python
reglasBienestar = {
    "estresado": "ğŸŒ¿ TÃ³mate 5 minutos para respirar profundo y desconectarte de la pantalla.",
    "cansado": "ğŸŒ¿ Haz una pausa breve y estira el cuerpo.",
    "aburrido": "ğŸ¨ Prueba una actividad creativa rÃ¡pida, como dibujar o escribir una idea.",
    "motivado": "âš¡ Aprovecha tu energÃ­a para avanzar en una tarea importante.",
    "feliz": "ğŸ˜Š Comparte tu energÃ­a con alguien mÃ¡s.",
    "desmotivado": "ğŸ’¡ Reorganiza tus tareas y comienza por algo sencillo.",
    "estres": "ğŸŒ¿ Respira profundo y desconÃ©ctate un momento.",
    "Con energÃ­a": "âš¡ Â¡Aprovecha el impulso para algo productivo!"
}
```

```python
class AgenteBienestar:
    def __init__(self, reglas):
        self.reglas = reglas

    def percibir(self, mensaje: str) -> list:
        # Convertir mensaje a minÃºsculas y separar palabras
        palabras = mensaje.lower().split()
        # Buscar percepciones en el diccionario
        percepciones = [p for p in self.reglas if p in palabras]
        return percepciones

    def decidir(self, percepciones: list) -> str:
        if percepciones:
            # Concatenar consejos en una frase mÃ¡s natural
            consejos = [self.reglas[p] for p in percepciones]
            if len(consejos) > 1:
                return "ğŸŒ± Consejo: " + ", pero tambiÃ©n ".join(consejos) + "."
            else:
                return "ğŸŒ± Consejo: " + consejos[0] + "."
        else:
            return "ğŸ’¡ Consejo: No comprendÃ­ tu estado, intenta describirlo con otra palabra."

    def conversar(self):
        print("ğŸ¤– Agente de Bienestar Digital Concatenado. Escribe cÃ³mo te sientes (o 'salir' para terminar).")
        while True:
            entrada = input("TÃº: ")
            if entrada.lower() == "salir":
                print("ğŸ¤– Agente: Â¡Gracias por cuidar tu bienestar digital! ğŸŒ±")
                break
            percepciones = self.percibir(entrada)
            respuesta = self.decidir(percepciones)
            print(f"ğŸ¤– Agente: {respuesta}")
```

```python
agente = AgenteBienestar(reglasBienestar)
agente.conversar()
```
